####  Насколько выводы по Xbox One коррелируют с продажами игр на других платформах

```python
top_platform_since_2013 = ['PS4', 'XOne', '3DS']

for i in top_platform_since_2013:
    df_for_corr = df[(df['platform'] == i) & (df['user_score'] > 0.1) & (df['critic_score'] > 0.1) & (df['year_of_release'] >= 2013)]
    user_data = df_for_corr['total_sales'].corr(df_for_corr['user_score'])
    critic_data = df_for_corr['total_sales'].corr(df_for_corr['critic_score'])
    
    if 0.01 < (user_data) < 0.29:
        enter_user = ('слабая положительная корреляция')
    elif 0.3 < (user_data) < 0.69:
        enter_user = ('средняя положительная корреляция')
    elif 0.7 < (user_data) <= 1:
        enter_user = ('сильная положительная корреляция')
    elif -0.29 < (user_data) < -0.01:
        enter_user = ('слабая отрицательная корреляция')
    elif -0.69 < (user_data) < -0.3:
        enter_user = ('средняя отрицательная корреляция')
    elif -1 < (user_data) <= -0.7:
        enter_user = ('сильная отрицательная корреляция')

    if 0.01 < (critic_data) < 0.29:
        enter_critic = ('слабая положительная корреляция')
    elif 0.3 < (critic_data) < 0.69:
        enter_critic = ('средняя положительная корреляция')
    elif 0.7 < (critic_data) <= 1:
        enter_critic = ('сильная положительная корреляция')
    elif -0.29 < (critic_data) < -0.01:
        enter_critic = ('слабая отрицательная корреляция')
    elif -0.69 < (critic_data) < -0.3:
        enter_critic = ('средняя отрицательная корреляция')
    elif -1 < (critic_data) <= -0.7:
        enter_critic = ('сильная отрицательная корреляция')
    print(i)
    print('Корреляция между пользовательскими отзывами и продажами: {:.2f} - {}'.format(user_data, enter_user))
    print('Корреляция между отзывами критиков и продажами: {:.2f} - {}'.format(critic_data, enter_critic))
    print('\n')
    
```

В целом, корреляция подтверждается и по остальным платформам.

Присутствуют платформы и с меньшим разбросом влияния между пользователями и критиками: 3DS

### Шаг 5. Проверка гипотезы

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

##### Нулевая гипотеза: 
средние пользовательские рейтинги платформ Xbox One и PC равны
    
##### Альтернативная гипотеза:
средние пользовательские рейтинги платформ Xbox One и PC не равны

    
Выборки у нас незавсимимые, поэтмоу используем метод scipy.stats.ttest_ind:

```python
from scipy import stats as st

#Формируем датасет / избавляемся от выбросов
df_xone = df['user_score'][(df['platform'] == 'XOne') & (df['year_of_release'] >= 2013)].dropna().reset_index()
df_pc = df['user_score'][(df['platform'] == 'PC') & (df['year_of_release'] >= 2013)].dropna().reset_index()

def quantiles(data, column):
    q1 = np.percentile(data[column].dropna(), 25)
    q3 = np.percentile(data[column].dropna(), 75)
    IQR = q3 - q1
    
    start = np.ceil(q1 - 1.5 * IQR)
    end = np.ceil(q3 + 1.5 * IQR)
    
    if start < 0:
        start = 0
    
    print('Значения интервала для {}: {} - {}'.format(column, start, end))
    return '\t'

print(quantiles(df_xone, 'user_score'))
print(quantiles(df_pc, 'user_score'))

df_xone = df_xone.query('user_score >= 3.0')
df_pc = df_pc.query('user_score >= 3.0')

#Проводим расчеты
alpha = 0.05

results = st.ttest_ind(df_xone['user_score'], df_pc['user_score'])
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Верно утверждение, что средние пользовательские рейтинги платформ Xbox One и PC не равны")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Верно утверждение, что средние пользовательские рейтинги платформ Xbox One и PC равны")
```

#### Cредние пользовательские рейтинги жанров Action и Sports разные

##### Нулевая гипотеза: 
средние пользовательские рейтинги жанров Action и Sports равны
    
##### Альтернативная гипотеза:
средние пользовательские рейтинги жанров Action и Sports не равны

    
Выборки у нас незавсимимые, поэтмоу используем метод scipy.stats.ttest_ind:

```python
#Формируем датасет / избавляемся от выбросов
df_action = df['user_score'][(df['genre'] == 'Action') & (df['user_score'] > 0.1) & (df['year_of_release'] >= 2013)].dropna().reset_index()
df_sport = df['user_score'][(df['genre'] == 'Sports') & (df['user_score'] > 0.1) & (df['year_of_release'] >= 2013)].dropna().reset_index()

#Проводим расчеты
alpha = 0.05

results = st.ttest_ind(df_action['user_score'], df_sport['user_score'])
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Врено утверждение, что средние пользовательские рейтинги жанров Action и Sports не равны")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Верно утверждение, что средние пользовательские рейтинги жанров Action и Sports равны")
```
